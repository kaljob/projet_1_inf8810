## Qu'elle est le contexte du jeu de données

Le site Inside Airbnb collecte et publie des données sur les annonces de locations touristiques à court terme sur Airbnb à travers le monde. Notre jeu de donnée represente des Annonce de location Airbnb Montreal à court terme. 
Un annonce represente une page sur Airbnb décrivant un logement disponible avec des détails comme la localisation, le prix, les équipements , les avis des précédents locataires et bien d'autres.

Ces informations sont collectées dans le but de données les informations à l'intéréssé (voyageur, touriste, etc) les informations nécessaire pour choisir où séjourner.

### Lecture du fichier (affichage des 10 premieres lignes)

```python
import pandas as pd
import patoolib
patoolib.extract_archive('listings.csv.gz', outdir='.')

data = pd.read_csv("./listings.csv", sep=",", header=0)
data.head(10)
```

### Pretraitement des données

```python
# Affiche la liste des colonnes
data.columns
```

```python
# Affiche le type des colonnes et le nombre de valeurs non nulles
data.info()
```

```python
# Ajout de 2 colonnes Ville et Pays. ces colonnes sont ajoutées à partir de la colonne 'host_location'
data['ville'] = data['host_location'].str.split(',').str[0]
data['pays'] = data['host_location'].str.split(',').str[1].str.strip()
```

```python
# Affiche les valeurs uniques de la colonne Pays
data['pays'].unique()
```

```python
# Nous Travaillerons uniquement sur les données du Canada
data_reduced = data[data['pays'] == 'Canada']
```

```python
# reduction des colonnes
data_cleanned = data_reduced[['name', 'description', 'neighborhood_overview', 'host_name', 'host_since', 'ville', 'property_type', 'room_type', 'accommodates', 'bathrooms',
       'bathrooms_text', 'bedrooms', 'beds', 'amenities', 'price',
       'minimum_nights', 'maximum_nights', 'number_of_reviews', 'review_scores_rating', 'last_review']]
```

```python
# Affiche les statistiques sur le jeu de données
data_cleanned.describe()
```

### Sauvegarde des notre jeu de donnée à utiliser pour les futurs traitements.

```python
data_cleanned.to_csv('listings_clean.csv')
```
